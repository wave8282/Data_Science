```python
import numpy as np
import matplotlib.pyplot as plt

mu1, sigma1 = 10, 2
mu2, sigma2 = -6, 3

standard =  np.random.randn(10000)
Gauss1 = mu1 + sigma1 *  np.random.randn(10000)
Gauss2 = mu2 + sigma2 *  np.random.randn(10000)

plt.figure(figsize=(10,6))
plt.hist(standard, bins=50, alpha=0.4)
plt.hist(Gauss1, bins=50, alpha=0.4)
plt.hist(Gauss2, bins=50, alpha=0.4)

plt.show()
```

```python
import matplotlib.pyplot as plt

month = ['7', '8', '9', '10', '11', '12']
users = [456, 492, 578, 599, 670, 854]

plt.plot(month, users)
plt.title("Daehan company speeda net new customers")
plt.xlabel("Month")
plt.ylabel("users")
plt.show()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(13,5))

ax[0].bar(month, users)
ax[0].set_title("Bar chart")
ax[0].set_xlabel("month")
ax[0].set_ylabel("users")
ax[1].plot(month, users)
ax[1].set_title("Line chart")
ax[2].plot(month, users, '^')
ax[2].set_title("Scatter chart")
ax[3].barh(month, users)
ax[3].set_title("Horizontal Bar chart")

plt.show
```

```python
fig = plt.figure(figsize=(12, 4))

plt.subplot(141)
plt.bar(month, users)
plt.title('Bar chart')
plt.subplot(142)
plt.plot(month, users)
plt.subplot(143)
plt.scatter(month, users)
plt.subplot(144)
plt.barh(month, users)
```
