```python
import pandas as pd
import matplotlib.pyplot as plt

weather = pd.read_csv('weather (3).csv', encoding='cp949')
weather['month'] = pd.DatetimeIndex(weather['일시']).month

months = range(1, 13)
monthly_wind = []

for m in months:
    month_df = weather[weather['month'] == m]
    monthly_wind.append(month_df['평균풍속'].mean())

plt.plot(months, monthly_wind)
plt.xlabel('month')
plt.ylabel('average wind speed(m/s)')
plt.show()
```

### concat() 함수로 데이터프레임을 합치기
### 데이터 병합 - merge() 함수

```python
import pandas as pd

df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                    'b':['b0','b1','b2','b3'],
                    'c':['c0','c1','c2','c3']},
                    index = [0, 1, 2, 3])

df2 = pd.DataFrame({'b':['b2','b3','b4','b5'],
                    'c':['c2','c3','c4','c5'],
                    'd':['d2','d3','d4','d5']},
                    index = [2, 3, 4, 5])
```

```python
df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                    'b':['b0','b1','b2','b3'],
                    'c':['c0','c1','c2','c3']},
                    index = [0, 1, 2, 3])

df2 = pd.DataFrame({'b':['b2','b3','b4','b5'],
                    'c':['c2','c3','c4','c5'],
                    'd':['d2','d3','d4','d5']},
                    index = [2, 3, 4, 5])
```

```python
df2
```

```python
df1
```

```python
pd.concat([df1, df2], join='outer') # 데이터를 모두 집어 넣을때
```

```python
pd.concat([df1, df2], join='inner') # 공통되는 칼럽만 집어 넣을 떄
```

```python
df1.merge(df2, how='outer', on='b')
```

```python
df1.merge(df2, how='inner', on='b')
```
