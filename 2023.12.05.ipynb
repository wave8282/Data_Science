### CSV(Comma-Separated Values)

```python
import csv

f = open('weather.csv')
data = csv.reader(f)
data = [field for field in data]
f.close()
```

```python
data
```

```python
header = data[0]
header
```

```python
data = data[1:]
data
```

```python
avg_wind_speed = [float(field[-1]) for field in data if field[-1]]
avg_wind_speed
```

```python
max(avg_wind_speed)
```

```python
month_speed = {}
for field in data:
    month = int(field[0].split('-')[1])
    if month not in month_speed:
        month_speed[month] = []
    month_speed[month].append(float(field[-1])) if field[-1] else 0.
    
month_avg_speed = {}
for k in sorted(month_speed):
    if k not in month_avg_speed:
        month_avg_speed[k] = sum(month_speed[k])/len(month_speed[k])
month_avg_speed

import matplotlib.pyplot as plt

plt.plot(month_avg_speed.keys(), month_avg_speed.values(), 'k-')
plt.xlabel('month')
plt.ylabel('avarage wind')
plt.show()
```

```python
 # weather['month'] = [int(date.split('-')[1])for date in weather['일시']]
weather['month'] = pd.DatetimeIndex(weather['일시']).month

avg_month_wind = []
for i in range(1, 13):
    avg_month_wind.append(weather[weather['month']==i]['평균풍속'].mean())
avg_month_wind

plt.plot(range(1, 13), avg_month_wind)
plt.show()
```
