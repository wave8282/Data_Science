```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] 
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(years, gdp, color='b', marker ='o', linestyle='solid', label="gdp of korea")

plt.title("GDP")
plt.ylabel("dollars")
plt.xlabel("years")
plt.legend() # 범례

plt.savefig("gdp_per_capita.png", dpi=150)
plt.show() # 출력 값 지우기
```

```python
x = np.arange(-20, 21)
y1 = 2 * x
y2 = x**2 + 5
y3 = - x**2 - 5

plt.plot(x, y1, 'r--', label='y=2x')
plt.plot(x, y2, 'b--', label='y=x^2+5')
plt.plot(x, y3, 'g--', label='y=-x^2-5')

plt.grid() # 영역
plt.xlim([-30, 30])  # 그림 그릴 영역 설정
plt.ylim([-30, 30])
plt.legend()
plt.show()
```

```python
import math 

theta = np.linspace(0, 360, 361)
sin = np.sin(2*np.pi*theta/360)
cos = np.cos(2*np.pi*theta/360)

plt.plot(theta, sin, 'k-', label='sin')
plt.plot(theta, cos, 'r--', label='cos')
plt.legend(loc=("lower left"))
plt.xlabel("theta")
plt.ylabel("values")
```

### 파이차트(pie chart)

### 히스토그램(histogram)

```python
books = [1,4,6,0,0,2,1,3]
plt.hist(books, bins = 6)
```

```python
plt.hist(x, bins = 20, alpha = 0.3) # bin 나누는 구간의 갯수
plt.hist(y, bins = 20, alpha = 0.3)
```

```python
n_bins = 20 
x = np.random.randn(1000) 
y = np.random.randn(1000) * 2

plt.hist(x, n_bins, histtype='bar', color='red',   label='std=1.0')
plt.hist(y, n_bins, histtype='bar', color='blue',  alpha=0.3, label='std=2.0')
plt.legend()
plt.show()
```

```python
rand1 = np.random.randn(1000)
rand2 = np.random.randn(1000)*2 + 10
rand3 = np.random.randn(1000)*3 + (-6)

plt.hist(rand1, 20, alpha=0.5)
plt.hist(rand2, 20, alpha=0.5)
plt.hist(rand3, 20, alpha=0.5)
plt.show()
```
